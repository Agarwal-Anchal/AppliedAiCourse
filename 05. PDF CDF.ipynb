```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv('iris.csv')
```

```python
iris_setosa=iris.loc[iris['target']==0.0]
iris_virginica=iris.loc[iris['target']==1.0]
iris_versicolor=iris.loc[iris['target']==2.0]
```

```python
plt.plot(iris_setosa['petal length (cm)'],np.zeros_like(iris_setosa['petal length (cm)']),'ro')
plt.plot(iris_virginica['petal length (cm)'],np.zeros_like(iris_virginica['petal length (cm)']),'go')
plt.plot(iris_versicolor['petal length (cm)'],np.zeros_like(iris_versicolor['petal length (cm)']),'bo')
plt.grid()
plt.show()

#can't differentiate between points hence histogram
```

```python
sns.FacetGrid(iris,hue='target',size=6) \
   .map(sns.distplot,'petal length (cm)') \
   .add_legend()
plt.grid()
plt.show()
```

```python
plt.hist(iris[iris['target']==0.0],bins=None)
```

```python
#CDF and PDF
#CDF can well separate any point for the histogram intersection
```

```python
counts,bin_edges=np.histogram(iris_setosa['petal length (cm)'],bins=10,density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
print(cdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.grid()
```
