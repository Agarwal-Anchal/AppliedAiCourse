```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv("iris.csv")
```

```python
iris_setosa=iris[iris['target']==0.0]
iris_virginica=iris[iris['target']==1.0]
iris_versicolor=iris[iris['target']==2.0]
```

```python
sns.boxplot(x='target',y='petal length (cm)',data=iris)
plt.grid()
plt.show()
```

```python
# no standard way to draw whiskers
# most points lie in whiskers
```

```python
#combine histogram and box plots
sns.violinplot(x='target',y='petal length (cm)',data=iris)
plt.grid()
```
